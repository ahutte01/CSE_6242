## NOAA Storm Events Data Processing

```python
import pandas as pd
import glob
import os
```

```python
storm_list = glob.glob(os.getcwd()+'/data/storm/*.csv')

for i,x in enumerate(storm_list):
    dat_i = pd.read_csv(x) #, usecols=['YEAR','MONTH_NAME','STATE', 'EVENT_TYPE', 'EPISODE_ID']
    dat_i = dat_i[dat_i.EVENT_TYPE.isin(['Coastal Flood'])].drop_duplicates()
    if i == 0:
        storm_dat = dat_i
    else:
        storm_dat = storm_dat.append(dat_i)
```

```python
storm_agg_for_model = storm_dat.groupby(['STATE', 'CZ_NAME']).nunique()
storm_agg_for_model['EVENT_ID'].to_csv("coastal_flood_events_by_county.csv") # Since 1996
storm_agg_for_model # After this, I did some manual editing in Excel
```

## Part 2: Taking the data I created by county and get it by zip

```python
storm_dat = pd.read_csv('coastal_flood_events_by_county.csv') # Manually edited csv file
county_zip = pd.read_csv('data/ZIP_COUNTY_122019.csv') # Zip to county mapping file
```

```python
storm_dat.head(), county_zip.head()
```

```python
merged = county_zip.merge(storm_dat, how = 'inner', left_on = 'COUNTY', right_on = 'County_FIPS')
```

```python
merged.describe()
```

```python
final = merged.drop(['COUNTY'], axis = 1)
```
